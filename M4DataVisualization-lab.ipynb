## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT count(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# sns.histplot(data=df, x='ConvertedComp')

# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 8))
sns.histplot(data=df, x='ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=df['Age'])
plt.xlabel('Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='Age', y='WorkWeekHrs')
plt.title('Age against Work Hours')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
plt.figure(figsize=(10, 8))
plt.rcParams['legend.fontsize'] = 25
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(10, 1000), alpha=0.5)
plt.title('Hours Worked Each Week against Code Review Hours')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python

QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)
databases = pd.DataFrame(df['DatabaseDesireNextYear'].value_counts()[0:4])
colors = sns.color_palette('pastel')[0:5]
labels = list(databases.index)

plt.pie(data=databases, x=databases['DatabaseDesireNextYear'], labels=labels, colors=colors, autopct='%.0f%%')
plt.show()
```

### Stacked Charts
`

Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df.head()

filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
medians = filtered_df[['WorkWeekHrs', 'CodeRevHrs']].groupby(filtered_df['Age']).median()
medians.plot(kind='bar', stacked=True, color=['red', 'skyblue'])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Hours')
plt.title('Median Hours Worked for the Age Range of 30 to 35')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
medians = filtered_df[['ConvertedComp']].groupby(filtered_df['Age']).median()
medians.plot(kind='line')
plt.ylabel('Compensation ($)')
plt.legend().remove()  # compensation increases with age, potentially due to experience/position? 
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
branch = pd.DataFrame(df['MainBranch'].value_counts())
sns.barplot(data=branch, x='MainBranch', y=branch.index, orient='h')
plt.xlabel('Mainbranch Count')
plt.title('Main Branch of Work') 
```
